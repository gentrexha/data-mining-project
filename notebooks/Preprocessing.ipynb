```python
import pandas as pd
import numpy as np

# Load everything
df_albin = pd.read_csv(
    "../data/raw/albini2017.csv",
    parse_dates=["Date"],
    infer_datetime_format=True,
    dtype={
        "Likes": np.int32,
        "Comments": np.int32,
        "Shares": np.int32,
    },
)
df_vjosa = pd.read_csv(
    "../data/raw/VjosaOsmaniMP.csv",
    parse_dates=["Date"],
    infer_datetime_format=True,
    dtype={
        "Likes": np.int32,
        "Comments": np.int32,
        "Shares": np.int32,
    },
)
df_kadri = pd.read_csv(
    "../data/raw/kadriveseliofficial.csv",
    parse_dates=["Date"],
    infer_datetime_format=True,
    dtype={
        "Likes": np.float64,
        "Comments": np.int32,
        "Shares": np.int32,
    },
)
df_ramush = pd.read_csv(
    "../data/raw/RamushHaradinajOfficial.csv",
    parse_dates=["Date"],
    infer_datetime_format=True,
    dtype={
        "Likes": np.int32,
        "Comments": np.int32,
        "Shares": np.int32,
    },
)
df_fatmir = pd.read_csv(
    "../data/raw/FatmirLimajKM.csv",
    parse_dates=["Date"],
    infer_datetime_format=True,
    dtype={
        "Likes": np.int32,
        "Comments": np.int32,
        "Shares": np.int32,
    },
)
```

```python
df_albin["Name"] = "Albin Kurti"
df_ramush["Name"] = "Ramush Haradinaj"
df_fatmir["Name"] = "Fatmir Limaj"
df_vjosa["Name"] = "Vjosa R. Osmani"
df_kadri["Name"] = "Kadri Veseli"
```

```python
df_fatmir_daily = df_fatmir.groupby(df_fatmir["Date"].dt.date)["Likes", "Shares", "Comments"].sum()
df_fatmir_daily["Posts"] = df_fatmir.groupby(df_fatmir["Date"].dt.date)["Date"].count()
df_fatmir_daily["Name"] = "Fatmir Limaj"
```
